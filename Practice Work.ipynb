```python
def clean_and_transform_data(df1, df2):
    # Remove unnecessary columns
    df1 = df1[['album','name', 'artists', 'duration_ms', 'popularity', 'explicit', 'preview_url']]
    df2 = df2[['album','name', 'artists', 'duration_ms', 'popularity', 'explicit', 'preview_url']]


    # Convert duration_ms to minutes:seconds

    def ms_to_min_sec(ms):
        minutes, seconds = divmod(ms / 1000, 60)
        return f"{int(minutes)}:{int(seconds)}"

    df1['duration_ms'] = df1['duration_ms'].apply(lambda x: ms_to_min_sec(x))
    df2['duration_ms'] = df2['duration_ms'].apply(lambda x: ms_to_min_sec(x))
    
    # Remove duplicate songs
    df1 = df1.drop_duplicates()
    df2 = df2.drop_duplicates()

    # Filter out songs with null preview_url
    df1 = df1[df1['preview_url'].notnull()]
    df2 = df2[df2['preview_url'].notnull()]

    # Rename and reorder columns
    df1 = df1.rename(columns={'name': 'song', 'duration_ms': 'duration'})
    df1 = df1[['song', 'artists', 'album', 'duration', 'popularity', 'explicit', 'preview_url']]

    df2 = df2.rename(columns={'name': 'song', 'duration_ms': 'duration'})
    df2 = df2[['song', 'artists','album', 'duration', 'popularity', 'explicit', 'preview_url']]

    # Combine the two DataFrames into a single DataFrame
    df = pd.concat([df1, df2])
    
    df1['artists'] = df1['artists'].str.replace("[\['\]']", "")

    return df
```

```python
def load_data_into_database(df):
    # Connect to the database
    conn = sqlite3.connect('mydatabase.db')

    # Write the DataFrame to a table in the database
    df.to_sql('playlists', conn, if_exists='replace')

    # Commit the changes
    conn.commit()

    # Close the connection
    conn.close()
```
